```python
plt.style.use('seaborn-v0_8')
```

```python
plt.plot(df.hitFlop )
plt.plot(df.sequel) 

plt.show()
```

```python
plt.style.available
```

```python
plt.figure(figsize=(7,7))
plt.style.use('Solarize_Light2')
plt.plot(df.hitFlop , color ='red' , marker = 'o' , label = 'hitFlop') 
plt.plot(df.sequel , label ="sequel" , color ="yellow" , marker ='*')
plt.style.use( 'seaborn-v0_8-notebook')
plt.title('sequel vs hitFlop')
plt.legend("hitFlop vs sequel ")

plt.show()
```

```python
df.head()
```

```python
plt.figure(figsize=(8,8))
plt.scatter(df.sequel, df.hitFlop , color ='c', marker='o',label= 'sequel')
plt.scatter(df.hitFlop ,df.sequel ,color ='y', marker='*',label= 'hitFlop') 
plt.xlabel('sequel')
plt.ylabel('hitFlop')
plt.title('sequel vs hitFlop' , color = 'b')
plt.legend()

plt.show()
```

```python
plt.style.use('seaborn-v0_8-dark')
plt.bar(df.sequel , df.hitFlop , color = 'yellow',  label= 'sequel')
plt.bar(df.hitFlop , df.sequel , color = 'red' , label= 'hitFlop')
plt.xlabel('sequel' , color = 'green')
plt.ylabel('hitFlop' , color = 'brown')
plt.title('sequel vs hitFlop' , color = 'blue')
plt.legend()
```
